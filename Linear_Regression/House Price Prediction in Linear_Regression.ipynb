```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_csv('homeprices.csv')
data
```

```python
plt.plot(data['area'],data['price'])
plt.show()
```

```python
area = data[['area']]
area
```

```python
price = data.price
price
```

```python
from sklearn.linear_model import LinearRegression
train = LinearRegression()
model = train.fit(area,price)
```

```python
model
```

```python
model.predict([[50200]])
```

```python
import pickle
with open('Linear_model_with_pickle','wb') as f:
    pickle.dump(model,f)
```
